```python
import pandas as pd

# Assuming 'sales_transaction_transactions_tbl' and 'sales_transaction_prod_cat_info_tbl' are your DataFrames

df = pd.merge(sales_transaction_transactions_tbl, 
              sales_transaction_prod_cat_info_tbl, 
              left_on='prod_subcat_code', 
              right_on='prod_sub_cat_code', 
              how='inner')
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Load your data
df = df #ansaction_transactions_tbl#pd.read_csv('your_data.csv')

# Assuming product_table is already defined
product_table = sales_transaction_prod_cat_info_tbl#pd.read_csv('product_table.csv')

# Perform one-hot encoding to convert the data into transaction format
basket_sets = df.groupby(['transaction_id', 'prod_subcat_code'])['Qty'].sum().unstack().reset_index().fillna(0).set_index('transaction_id')

# Convert quantities to binary values (1 if item was purchased, 0 otherwise)
basket_sets = basket_sets.applymap(lambda x: 1 if x > 0 else 0)

# Apply Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(basket_sets, min_support=0.1, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Add a new column in the product table to indicate association rule results
product_table['association_rule'] = ''  # Initialize column
for index, row in rules.iterrows():
    antecedents = set(row['antecedents'])
    consequents = set(row['consequents'])
    for prod_subcat_code in product_table['prod_subcat_code']:
        if antecedents.issubset({prod_subcat_code}) or consequents.issubset({prod_subcat_code}):
            product_table.loc[product_table['prod_subcat_code'] == prod_subcat_code, 'association_rule'] = ' '.join(row['antecedents']) + '->' + ' '.join(row['consequents'])

# Save the modified product table
product_table.to_csv('product_table_with_association.csv', index=False)
```
